```python
import pandas as pd
import os

loading_student_data = os.path.join("Resources",
                                    "students_complete.csv")

reading_student_data = pd.read_csv(loading_student_data)
reading_student_data
```

```python
student_name_list = reading_student_data["student_name"].tolist()
student_name_list
```

```python
name_list_of_len3plus = []
for name in student_name_list:
    name_split = name.split()
    name_length = len(name_split)
    print(f'Name: {name_split}; length: {name_length}')
```

```python
name_list_of_len3plus = []
for name in student_name_list:
    name_split = name.split()
    name_length = len(name_split)
    if name_length >= 3:
        name_list_of_len3plus.append(name)
        # NOTE hat having print(name_list_of_len3plus) here will result in printing a new line each time there
        # is a new name_split with len greater than 3
    # NOTE that having print(name_list_of_len3plus) here will result in printing a new line for each iteration
print(name_list_of_len3plus)
```

```python
len(name_list_of_len3plus)
```

```python
# PREFIX --> [0]
prefix_list = []

for index in name_list_of_len3plus:
    if len(index.split()[0]) <= 4:
        # index.split() --> list of index AFTER split
        # if the length of the individual split is 4 or greater then add the first
        # position of the split into the new list made --> NOTE 4 because the longest
        # known prefix in the list is 4
        prefix_list.append(index.split()[0])

prefix_list
```

```python
# SUFFIX --> [-1]
suffix_list = []

for index in name_list_of_len3plus:
    if len(index.split()[-1]) <= 3:
        # if the length of the split is 3 or greater then add the last position of the
        # split into the new list made --> NOTE 3 because the longest known suffix in 
        # the list is 3
        suffix_list.append(index.split()[-1])

suffix_list
```

```python
set(prefix_list)
    # obtains the unique items in prefix_list --> eliminates duplicates
```

```python
set(suffix_list)
    # obtains the unique items in suffix_list --> eliminates duplicates
```

```python
prefix_and_suffix_list = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
    # NOTE that this list was manually completed based on the set of the prefix_list and suffix_list
    # NOTE that there are spaces to account for whitespaces once they are removed
    
for word in prefix_and_suffix_list:
    reading_student_data["student_name"] = reading_student_data["student_name"].str.replace(word,"")
reading_student_data
```

```python
# Creating a list for student names with still a length of 3 or greater

semi_fixed_student_name = reading_student_data["student_name"].tolist()
semi_fixed_student_name_list = []
for name in semi_fixed_student_name:
    name_split = name.split()
    name_length = len(name_split)
    if name_length >= 3:
        semi_fixed_student_name_list.append(name)
print(f' {semi_fixed_student_name_list} \n with a total of {len(semi_fixed_student_name_list)}')
```
