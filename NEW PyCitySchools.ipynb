```python
student_name_with_error = reading_student_data["student_name"]
# PREFIX --> [0]
prefix_list = []

for index in student_name_with_error:
    if len(index.split()) >= 4:
        # if the length of the row split is equal to or greater than 4 append the first index to prefix_list
        # NOTE that this is risky because this could not account for rows less then length 4 that has a 
        # prefix to replace
        prefix_list.append(index.split()[0])

set(prefix_list)
```

```python
# suffix --> [-1]
suffix_list = []

for index in student_name_with_error:
    if len(index.split()) >= 4:
        # if the length of the row split is equal to or greater than 4 append the first index to suffix_list
        # NOTE that this is risky because this could not account for rows less then length 4 that has a 
        # suffix to replace
        suffix_list.append(index.split()[-1])

set(suffix_list)
```

```python
prefix_and_suffix_list = ['Dr. ', 'Miss ', 'Mr. ', 'Mrs. ', 'Ms. ', ' DDS', ' DVM', ' MD', ' PhD']
    # NOTE that the spaces are to account for the whitespace that would be left over when deleing the
    # prefix/suffix
for prefix_or_suffix in prefix_and_suffix_list:
   reading_student_data["student_name"] = reading_student_data["student_name"].str.replace(prefix_or_suffix,"")
    
reading_student_data
```
